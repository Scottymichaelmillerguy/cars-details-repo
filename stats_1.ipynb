# Car Dataset

## Intro

```python
cars = pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")
cars_df = pd.DataFrame(cars)

cars_df.head()
```

```python
cars_df.info()
```

```python
cars_df.describe()
```

```python
#checking for null values
cars_df.isnull().sum()
```

```python
car_name = cars_df.name
cars_df['make'] = cars_df.name.apply(lambda x: ''.join(x.split(' ')[0]))
cars_df['model'] = cars_df.name.apply(lambda x: ''.join(x.split(' ')[1:]))


display(cars_df.head(5))
```

## Descriptive Statistics

```python
selling_price_mean = cars_df.selling_price.mean()
selling_price_min = cars_df.selling_price.min()
selling_price_max = cars_df.selling_price.max()

print('Mean Selling Price:', selling_price_mean)
print('Min Selling Price:', selling_price_min)
print('Max Selling Price:', selling_price_max)

cars_df["# standard deviations from mean selling price"] = abs(cars_df["selling_price"] - selling_price_mean)/cars_df.selling_price.std()

km_mean = cars_df.km_driven.mean()
km_min = cars_df.km_driven.min()
km_max = cars_df.km_driven.max()

print('Mean Distance Driven (km):', km_mean)
print('Min Distance Driven (km):', km_min)
print('Max Distance Driven (km):', km_max)

cars_df["# standard deviations from mean distance driven"] = abs(cars_df["km_driven"] - km_mean)/cars_df.km_driven.std()

display(cars_df)
```

## Car Variables

```python
car_name = cars_df['name']
car_year = cars_df['year']
car_price = cars_df['selling_price']
car_distance = cars_df['km_driven']
car_fuel = cars_df['fuel']
car_seller = cars_df['seller_type']
car_transmission = cars_df['transmission']
car_owner = cars_df['owner']
car_make = cars_df['make']
car_model = cars_df['model']
```

### Selling Price Cluster Analysis

#### Numerical Variables

##### Year vs Price

```python
df_year_price = cars_df[['year', 'selling_price']]
display(df_year_price)
kmeans = KMeans(n_clusters = 3).fit(df_year_price)
```

```python
df_year_price['cluster'] = kmeans.labels_
df_year_price
```

```python
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=['year', 'selling_price'])
centroids
```

```python
fig, yp = plt.subplots(figsize=(5,5))
yp = sb.scatterplot(x='year', y='selling_price', hue='cluster', style='cluster', data=df_year_price)
centroids.plot.scatter(x='year', y='selling_price', ax=yp, s=50, color='red')

plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.title('KMeans Clustering: Year vs Selling Price')
plt.legend()
plt.grid(True)
plt.show()
```

##### Distance Driven vs Price

```python
df_distance_price = cars_df[['km_driven', 'selling_price']]
display(df_distance_price)
kmeans = KMeans(n_clusters = 3).fit(df_distance_price)
```

```python
df_distance_price['cluster'] = kmeans.labels_
df_distance_price
```

```python
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=['km_driven', 'selling_price'])
centroids
```

```python
fig, dp = plt.subplots(figsize=(5,5))
dp = sb.scatterplot(x='km_driven', y='selling_price', hue='cluster', style='cluster', data=df_distance_price)
centroids.plot.scatter(x='km_driven', y='selling_price', ax=dp, s=50, color='blue')

plt.xlabel('Distance Driven (KM)')
plt.ylabel('Selling Price')
plt.title('KMeans Clustering: Distance Driven vs Selling Price')
plt.legend()
plt.grid(True)
plt.show()
```
